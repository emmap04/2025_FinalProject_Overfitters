```python
import pandas as pd
divorce = pd.read_csv('data/Marriage_Divorce_DB.csv')
```

```python
divorce.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression             
from sklearn.model_selection import train_test_split      
from sklearn.linear_model import LinearRegression         
from sklearn.metrics import mean_squared_error, r2_score 
```

```python
## finding the most correlated 
r_scores = {}
for col in divorce.columns[:-1]:
    x = divorce[[col]]
    y = divorce['Divorce Probability']

    model = LinearRegression()
    model.fit(x, y)

    y_pred = model.predict(x)
    r2 = r2_score(y, y_pred)
    r_scores[col] = r2
r_scores

sorted_rscore = sorted(r_scores.items(), key=lambda item: item[1], reverse=True)
sorted_rscore
```
